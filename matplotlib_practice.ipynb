```python
# To make sure our plots are displayed inline within the Jupyter Notebook directly below the code cell that produced it
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add label for the legend.
plt.plot(x_axis, y_axis, label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
# Change the line color to green, the marker to diamond, line width to 2 pt, legend for Boston.
ax.plot(x_axis, y_axis, label="Boston", color="green", marker="d", linewidth=2)
# Add a title
ax.set_title("PyBer Fare by Month")
# Add x and y axes labels
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
# Add a y axis limit
ax.set_ylim(0, 45)
# Add grid lines
ax.grid()
# Add legend
ax.legend()
```

```python
# Bubble Chart
```

```python
# Create the plot
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Iterate through y-axis values to make them bigger
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

# Create the plot
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Using the list comprehension technique to multiply each data point in the y-axis list, 
# we would do the following in the plt.scatter() function:
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Skill drill
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], c="xkcd:sky blue", alpha=0.2, edgecolors="black", linewidths=2, label="Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0, 50)
ax.invert_yaxis()
ax.legend()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
    shadow=True,
    startangle=90)
ax.invert_xaxis()
```
